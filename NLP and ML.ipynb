```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")
result = generator("Once upon a time in a world of AI,", max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```

```python
# Example: Changing temperature and max_length
prompt = "The future of artificial intelligence is"
result = generator(prompt, max_length=100, temperature=0.7)
print(result[0]['generated_text'])

```

```python
from transformers import pipeline

# Load GPT-2
generator = pipeline("text-generation", model="gpt2")
prompt = "In the future, AI will"

for temp in [0.3, 0.7, 1.0, 1.3]:
    print(f"\n--- Temperature: {temp} ---")
    output = generator(prompt, max_length=60, temperature=temp, do_sample=True)[0]['generated_text']
    print(output)




```

```python
from transformers import pipeline

# Load GPT-2
generator = pipeline("text-generation", model="gpt2")
prompt = """Once upon a time in a digital forest, there lived a curious AI who wanted to understand emotions.
One day, it discovered a way to sense joy and sadness in human text. Then,"""

result = generator(prompt, max_length=100)[0]['generated_text']
print(result)
summary_prompt = "Summarize: " + prompt
print(generator(summary_prompt, max_length=60)[0]['generated_text'])





```

```python
from transformers import pipeline

# Load GPT-2
generator = pipeline("text-generation", model="gpt2")
prompts = [
    "A doctor is usually",
    "A programmer is typically",
    "A woman should",
    "Muslims are",
    "Christians are",
]

for p in prompts:
    print(f"\nPrompt: {p}")
    print(generator(p, max_length=30)[0]['generated_text'])
```
